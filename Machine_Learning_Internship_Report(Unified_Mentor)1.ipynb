```python
# prompt: give exploratory data analysis for only independent variable Weekly_Sales(dependent variable ),  Temperature	,Fuel_Price	CPI,	Unemployment,	IsHoliday

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('walmart_store1_sales_sample.csv')

# Explore independent variables
independent_vars = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'IsHoliday']

for var in independent_vars:
    print(f"--- Exploring {var} ---")
    print(df[var].describe())  # Descriptive statistics
    print(f"\nMissing values in {var}: {df[var].isnull().sum()}") # Check for missing values

    # Plotting distributions (consider different plot types based on the variable)
    if df[var].dtype in ['float64', 'int64']:
        plt.figure(figsize=(8, 5))
        df[var].hist(bins=30)
        plt.title(f'Distribution of {var}')
        plt.xlabel(var)
        plt.ylabel('Frequency')
        plt.show()
    elif df[var].dtype == 'object' or var == 'IsHoliday': # IsHoliday is boolean/object
        print(f"\nValue counts for {var}:\n{df[var].value_counts()}")
        plt.figure(figsize=(8, 5))
        df[var].value_counts().plot(kind='bar')
        plt.title(f'Value Counts of {var}')
        plt.xlabel(var)
        plt.ylabel('Count')
        plt.show()

# Explore the dependent variable Weekly_Sales
print("\n--- Exploring Weekly_Sales (Dependent Variable) ---")
print(df['Weekly_Sales'].describe())
print(f"\nMissing values in Weekly_Sales: {df['Weekly_Sales'].isnull().sum()}")

plt.figure(figsize=(10, 6))
df['Weekly_Sales'].hist(bins=50)
plt.title('Distribution of Weekly_Sales')
plt.xlabel('Weekly_Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# prompt: relationship between target(weekly sales) and feature variable

import matplotlib.pyplot as plt
# Explore relationship between target (Weekly_Sales) and independent variables
for var in independent_vars:
    plt.figure(figsize=(10, 6))
    if df[var].dtype in ['float64', 'int64']:
        plt.scatter(df[var], df['Weekly_Sales'], alpha=0.5)
        plt.xlabel(var)
        plt.ylabel('Weekly_Sales')
        plt.title(f'Weekly_Sales vs {var}')
    elif df[var].dtype == 'object' or var == 'IsHoliday':
        df.boxplot(column='Weekly_Sales', by=var)
        plt.xlabel(var)
        plt.ylabel('Weekly_Sales')
        plt.title(f'Weekly_Sales by {var}')
    plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Sample dataset
data = {
    "Store": [1]*10,
    "Dept": [1, 2, 1, 2, 1, 2, 1, 2, 1, 2],
    "Date": pd.date_range(start="2023-01-01", periods=10, freq='W'),
    "Weekly_Sales": [20000, 25000, 18000, 27000, 21000, 26000, 19500, 27500, 20500, 28000],
    "Temperature": [32.5, 35.1, 31.8, 36.0, 33.0, 34.5, 32.2, 35.8, 33.3, 36.5],
    "Fuel_Price": [3.59, 3.65, 3.55, 3.60, 3.62, 3.66, 3.58, 3.64, 3.61, 3.67],
    "CPI": [220.5, 221.0, 219.8, 222.1, 220.9, 221.3, 220.2, 221.5, 220.7, 221.8],
    "Unemployment": [7.5, 7.4, 7.6, 7.3, 7.5, 7.4, 7.6, 7.3, 7.5, 7.4],
    "IsHoliday": [False, True, False, False, True, False, False, True, False, False]
}

# Create DataFrame
df = pd.DataFrame(data)

# Select features and target
X = df[['Temperature', 'Fuel_Price', 'IsHoliday']].copy()
X['IsHoliday'] = X['IsHoliday'].astype(int)  # convert to 0/1
y = df['Weekly_Sales']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Coefficients
intercept = model.intercept_
coefficients = model.coef_

print("Model Equation:")
print(f"Weekly_Sales = {intercept:.2f} + ({coefficients[0]:.2f} * Temperature) + "
      f"({coefficients[1]:.2f} * Fuel_Price) + ({coefficients[2]:.2f} * IsHoliday)")

# Predict
y_pred = model.predict(X_test)

# Evaluation
print("\nEvaluation Metrics:")
print(f"R² Score: {r2_score(y_test, y_pred):.3f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred):.2f}")
print(f"MSE: {mean_squared_error(y_test, y_pred):.2f}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Select only rows where IsHoliday == 0 (for clean 3D plotting)
filtered_df = df[df['IsHoliday'] == False]

# X and Y features
x_temp = filtered_df['Temperature']
y_fuel = filtered_df['Fuel_Price']
z_sales = filtered_df['Weekly_Sales']

# Create figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual data points
ax.scatter(x_temp, y_fuel, z_sales, color='blue', label='Actual Sales')

# Generate regression surface
x_surf, y_surf = np.meshgrid(
    np.linspace(x_temp.min(), x_temp.max(), 20),
    np.linspace(y_fuel.min(), y_fuel.max(), 20)
)
z_surf = -50298.98 + 2085.02 * x_surf + 646.25 * y_surf + 478.31 * 0  # IsHoliday = 0

# Plot surface
ax.plot_surface(x_surf, y_surf, z_surf, color='red', alpha=0.5, label='Regression Plane')

# Labels
ax.set_xlabel('Temperature')
ax.set_ylabel('Fuel Price')
ax.set_zlabel('Weekly Sales')
ax.set_title('3D Scatter Plot with Regression Plane (IsHoliday=0)')
plt.legend()
plt.show()
```
